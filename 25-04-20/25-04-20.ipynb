```python
# Rule 22 tinkering

import numpy as np
import matplotlib.pyplot as plt

def run_1d_ca(initial_state, steps):
    # Initialize the state
    state = np.array(initial_state)
    history = [state.copy()]
    
    for _ in range(steps):
        # Create a new state array to store the updated values
        new_state = state.copy()
        
        # Apply the rule: cell becomes 1 if it has exactly one neighboring 1
        for i in range(1, len(state) - 1):  # skip the boundary cells
            if state[i] == 1:
                # Check neighbors for exactly one '1'
                if (state[i - 1] == 1 and state[i + 1] == 0) or (state[i - 1] == 0 and state[i + 1] == 1):
                    new_state[i] = 1
                else:
                    new_state[i] = 0
            else:
                # Cell is 0, apply rule to see if it becomes 1
                if (state[i - 1] == 1 and state[i + 1] == 0) or (state[i - 1] == 0 and state[i + 1] == 1):
                    new_state[i] = 1
                else:
                    new_state[i] = 0
        
        # Add the new state to history
        state = new_state
        history.append(state.copy())
    
    return np.array(history)

def plot_ca_history(history):
    plt.figure(figsize=(10, 6))
    plt.imshow(history, cmap='binary', interpolation='nearest', aspect='auto')
    plt.title("1D Cellular Automaton Evolution")
    plt.xlabel("Cell Index")
    plt.ylabel("Generation")
    plt.colorbar(label='Cell State (1 = Alive, 0 = Dead)')
    plt.show()

# Initialize the 1D cellular automaton with the initial state and steps
initial_state = [1] + [0] * 499
steps = 500  # Number of steps to run the CA

# Run the cellular automaton and visualize the behavior
history = run_1d_ca(initial_state, steps)
plot_ca_history(history)
```
