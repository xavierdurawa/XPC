```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import numpy as np
from matplotlib.animation import PillowWriter

# Parameters for the grid
rows, cols = 10, 3
dot_color = 'blue'
grid_color = 'lightgray'
filled_color = 'red'

# Create the figure and axes
fig, ax = plt.subplots(figsize=(6, 10))  # Increase the width to allow space for labels
ax.set_xlim(-1.5, cols + 0.5)  # Adjust limits to include "Window" and "Aisle"
ax.set_ylim(-0.5, rows - 0.5)
ax.set_xticks(np.arange(-0.5, cols, 1), minor=True)
ax.set_yticks(np.arange(-0.5, rows, 1), minor=True)
ax.grid(which="minor", color=grid_color, linestyle='-', linewidth=2)
ax.set_xticks([])
ax.set_yticks([])

# Add labels for "Front" and "Back"
ax.text(1.5, rows, 'Front', fontsize=16, ha='center', va='bottom', color='black')
ax.text(1.5, -1, 'Back', fontsize=16, ha='center', va='top', color='black')

# Add labels for "Window" and "Aisle"
ax.text(-1.5, rows / 2, 'Window', fontsize=14, ha='center', va='center', color='black', rotation=90)
ax.text(cols + 0.5, rows / 2, 'Aisle', fontsize=14, ha='center', va='center', color='black', rotation=90)

# Create the dots
dots = []
for i in range(cols):
    col_dots = []
    for j in range(rows):
        dot, = ax.plot(i, j, 'o', color=dot_color, markersize=10)
        col_dots.append(dot)
    dots.append(col_dots)

# Function to update the animation
def update(frame):
    col = frame // rows
    row = frame % rows
    if col < cols:
        dots[col][row].set_color(filled_color)

# Number of frames in the animation
num_frames = rows * cols

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=num_frames, interval=500, repeat=False)

# Save the animation as a GIF
gif_path = "dots_animation_with_window_aisle_fixed.gif"
ani.save(gif_path, writer=PillowWriter(fps=2))  # Save with 2 frames per second

print(f"Animation saved as {gif_path}")
```
